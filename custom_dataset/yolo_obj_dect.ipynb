```python
import os
import traceback
from simple_image_download import simple_image_download as simp

response = simp.Downloader

keywords = ["steak", "fried-chicken", "donut", "rice", "lettuce", "noodles", "french-fries", "ice-cream", "cauliflower", "fried-egg", "bacon", "pancake", "waffle", "cheese", "cookies"]

pre_dataset = os.path.join(os.getcwd(), "pre-dataset")

if not os.path.exists(os.path.join(pre_dataset, "images")):
    os.makedirs(os.path.join(pre_dataset, "images"))

for keyword in keywords:
    try:
        response().download(keyword, limit=40, verbose=True, cache=False, download_cache=False)
    except:
        print(f"Error downloading {keyword}")
        traceback.print_exc()

    for file in os.listdir(os.path.join(os.getcwd(), "simple_images", keyword)):
        os.rename(os.path.join(os.getcwd(), "simple_images", keyword, file), os.path.join(pre_dataset, "images", file))
```

### Delete unlabeled images

```python
import os

img_list = os.listdir("pre-dataset/images")
lab_list = os.listdir("pre-dataset/labels")

# remove if there is no label file .txt with the same name as the image, but image can be with any extension like .jpg, .png, .jpeg
for img in img_list:
    if img.split(".")[0] + ".txt" not in lab_list:
        os.remove("pre-dataset/images/" + img)
        print(f"removed {img}")
```

### Creating Train and Validation Dataset

```python
import os
import shutil
import random

# move classes.txt to dataset folder
if not os.path.exists("dataset"):
    os.makedirs("dataset")

#  Copy classes.txt to dataset folder
shutil.copy("pre-dataset/labels/classes.txt", "dataset/classes.txt")



img_list = os.listdir("pre-dataset/images")
lab_list = os.listdir("pre-dataset/labels")

grouped_img_list = []

for keyword in keywords:
    grouped_img_list.append([img for img in img_list if keyword in img])

train_path = "dataset/train"
val_path = "dataset/val"

if not os.path.exists(train_path):
    os.makedirs(train_path)
    if not os.path.exists(train_path + "/images"):
        os.makedirs(train_path + "/images")
    if not os.path.exists(train_path + "/labels"):
        os.makedirs(train_path + "/labels")


if not os.path.exists(val_path):
    os.makedirs(val_path)
    if not os.path.exists(val_path + "/images"):
        os.makedirs(val_path + "/images")
    if not os.path.exists(val_path + "/labels"):
        os.makedirs(val_path + "/labels")

val_split = 0.2

for imgs in grouped_img_list:
    random.shuffle(imgs)
    for img in imgs[:int(len(imgs) * val_split)]:
        shutil.copy("pre-dataset/images/" + img, val_path + "/images/" + img)
        shutil.copy("pre-dataset/labels/" + img.split(".")[0] + ".txt", val_path + "/labels/" + img.split(".")[0] + ".txt")

    for img in imgs[int(len(imgs) * val_split):]:
        shutil.copy("pre-dataset/images/" + img, train_path + "/images/" + img)
        shutil.copy("pre-dataset/labels/" + img.split(".")[0] + ".txt", train_path + "/labels/" + img.split(".")[0] + ".txt")
```
